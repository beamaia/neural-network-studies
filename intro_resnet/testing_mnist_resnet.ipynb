```python
import resnet
import torch
from torchvision import models
from mnist_resnet import resnet50, MnistResNet
import mnist_dataset as mnist
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
train_loader, test_loader = mnist.create_DataLoader() 
```

```python
model = MnistResNet() 
resnet.train_model(model, train_loader, device) 
```

```python
model.eval()
model.to(device)
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in test_loader:
        images = images.resize_(100, 1, 32, 32).to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print('Accuracy of the model on the test images: {} %'.format(100 * correct / total))
```
